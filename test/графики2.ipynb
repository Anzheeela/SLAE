```python
df = pd.read_csv('iterations_and_time.csv', sep = '\t')
plt.figure()
plt.plot(df['eps'], df['Gauss-Seidel, time'], '.', color = 'blue', label = 'Гаусс-Зейдель')
plt.plot(df['eps'], df['Jacobi, time'], '.', color = 'red', label = 'Якоби')
plt.xlabel('точность')
plt.ylabel('время выполнения программы')
plt.legend()

plt.figure()
plt.plot(df['eps'], df['Gauss-Seidel, iterations'], '.', color = 'blue', label = 'Гаусс-Зейдель')
plt.xlabel('точность')
plt.ylabel('количество итераций')
plt.legend()

display(df)
```
